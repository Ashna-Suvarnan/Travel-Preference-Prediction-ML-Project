# **TRAVEL PREFERENCE PREDICTION**

# **Objective Of The Study**



*   This dataset aims to analyze public preferences between two popular vacation types: mountains and beaches.

*    It provides insights into various personal and social factors that may influence these preferences.


*   This can be useful for travel agencies, tourism businesses, and individuals looking for destination suggestions.


*   The Preference variable represents an individual's vacation choice, where 0 indicates a preference for mountains and 1 indicates a preference for beaches.



# Features


Age: Age of the individual

Gender: Gender identity of the individual

Income: Annual income of the individual

Education Level: Highest level of education attained

Travel Frequency: Number of vacations taken per year

Vacation Budget: Budget allocated for vacations

Location: Type of residence

Proximity to Mountains: Distance from the nearest mountains

Proximity to Beaches: Distance from the nearest beaches

Pets: Indicates whether the individual owns pets (binary: 0 = No, 1 = Yes).

Environmental Concerns: Indicates whether the individual has environmental concerns (binary: 0 = No, 1 = Yes).

```python

df.drop(['Gender','Location'],axis=1,inplace=True)
```

# **MODEL CREATION**

# Data Splitting And Future Scaling

```python
x = df.drop('Preference',axis = 1)
y = df['Preference']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20 , random_state =42)
```

```python
from sklearn.preprocessing import StandardScaler

sd = StandardScaler()
sd.fit(x_train)
x_train = sd.transform(x_train)
x_test = sd.transform(x_test)
```

# **Performance Evaluation of Classification Models with Hyperparameter Tuning and SMOTE**

# 1) K-NEAREST NEIGHBOR (KNN)

```python

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python

param_dist = {'n_neighbors':[3, 5, 7,11, 15],'metric': ['euclidean', 'manhattan', 'minkowski'],'weights': ['uniform', 'distance'],'p': [1, 2]}

from sklearn.model_selection import RandomizedSearchCV

random_search = RandomizedSearchCV(knn, param_dist, n_iter=10, cv=5, random_state=42)
random_search.fit(x_train, y_train)
```

```python

print("Best Parameters:", random_search.best_params_)
print("Best Score:",random_search.best_score_ )
```

```python
best_knn = KNeighborsClassifier(weights='distance', p=1, n_neighbors=15, metric='minkowski')
best_knn.fit(x_train,y_train)
```

```python
y_pred1 = best_knn.predict(x_test)
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy = accuracy_score(y_test, y_pred1)
print('Accuracy: ', accuracy * 100, '\n')
print(confusion_matrix(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
xres,yres = sm.fit_resample(x_train,y_train)
```

```python
yres.value_counts()
```

```python
best_knn_smote = KNeighborsClassifier(weights='distance', p=1, n_neighbors=15, metric='minkowski')
best_knn_smote.fit(xres,yres)
```

```python
y_pred1 = best_knn_smote.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy = accuracy_score(y_test, y_pred1)
print('Accuracy: ', accuracy * 100, '\n')
print(confusion_matrix(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```
